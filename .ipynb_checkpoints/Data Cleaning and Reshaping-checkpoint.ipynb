```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
from matplotlib.pyplot import figure
matplotlib.style.use('fivethirtyeight')
```

```python
mvps = pd.read_csv("./files/mvps.csv")
mvps.head()
```

# Create a subset of MVPs dataset by extracting only those we need to join to the players dataset

```python
#Pts Won - Points earned for MPV voting
#Pts Max - Maximum amount of points possible
#Share - Percentage of points wons
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps
```

```python
players = pd.read_csv("./files/player_per_game.csv")
players.head()
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

# Merge the MVPs and Players datasets on Player Name and Year

## Upon inspecting the player's column, we have to remove the asterick in order the join to be successful

```python
players["Player"].head(50)
```

```python
players['Player'] = players['Player'].str.replace("*", "", regex= False)
```

```python
players.head(50)
```

## Another issue is that a player will have multiple records if they played on different teams. We want only one row per player for our analysis

```python
players.groupby(["Player", "Year"]).get_group(("Tariq Abdul-Wahad", 2000))
```

```python
'''This function take in a single dataframe and returns the record if there is only one row. 
If there are multiple rows for a player, it will take the total or TOT and replace the Team(Tm) with the 
most current team.
'''

def single_record(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] == df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_record)
```

```python
players.head(25)
```

## We created a multilevel index with the group by operations so we will need to drop the index level twice 

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
players[players["Tm"] == "TOT"]
```

```python
#Confirm that we have successfully combined each player into one record per year

players[players["Player"] == "Tariq Abdul-Wahad"]
```

```python
combined_df = players.merge(mvps, how="outer", on = ["Player", "Year"])
combined_df.head(20)
```

```python
combined_df[combined_df["Pts Won"] > 0]
```

## Fill in the NaN values under Pts Won, Pts Max, and Share with 0

```python
combined_df[["Pts Won", "Pts Max", "Share"]] = combined_df[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined_df
```

```python
teams = pd.read_csv("./files/team_standings.csv")
teams.head(20)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(20)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex = False)
teams.head(20)
```

```python
teams["Team"].unique()
```

```python
combined_df["Tm"].unique()
```

```python
nicknames = pd.read_csv("./files/team_nicknames.csv")
nicknames.head(10)
```

```python
nicknames = {}

with open("./files/team_nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
        
nicknames
```

```python
combined_df["Team"] = combined_df["Tm"].map(nicknames)
combined_df
```

```python
stats_df = combined_df.merge(teams, how = "outer", on = ["Team", "Year"])

stats_df
```

```python
del stats_df["Unnamed: 0"]
```

```python
stats_df.dtypes
```

```python
stats_df = stats_df.apply(pd.to_numeric, errors = "ignore")
stats_df.dtypes
```

```python
stats_df["GB"].unique()
```

```python
stats_df["GB"] = stats_df["GB"].str.replace("—", "0")
stats_df["GB"].unique()
```

```python
stats_df["GB"] = pd.to_numeric(stats_df["GB"])
stats_df.dtypes
```

# Export the final dataset

```python
stats_df.to_csv("./files/player_mvp_stats.csv")
```

# Exploratory Data Analysis

```python
highest_scoring = stats_df[stats_df["G"]> 70].sort_values("PTS", ascending = False).head(10)
highest_scoring
```

```python
x_labels = list(highest_scoring["PTS"])
```

```python
ax = highest_scoring.plot.barh("Player", "PTS")


ax.set_title("Top 10 Scorers from 2000 - 2021")
ax.set_ylabel("Player Name")
ax.set_xlabel("Points")
ax.get_legend().remove()
for i, v in enumerate(x_labels):
    ax.text(v + .75, i - 0.25, str(v), color='royalblue')
```

```python
highest_dimes = stats_df[stats_df["G"]> 70].sort_values("AST", ascending = False).head(10)
highest_dimes
```

```python
x_labels = list(highest_dimes["AST"])

ax = highest_dimes.plot.barh("Player", "AST")

ax.set_title("Top 10 Assist Leaders from 2000 - 2021")
ax.set_ylabel("Player Name")
ax.set_xlabel("Assist")
ax.grid(False)

ax.get_legend().remove()

for i, v in enumerate(x_labels):
    ax.text(v + .25, i - 0.25, str(v), color='royalblue')
```

```python
highest_boards = stats_df[stats_df["G"]> 70].sort_values("TRB", ascending = False).head(10)
highest_boards
```

```python
x_labels = list(highest_boards["TRB"])
```

```python

ax = highest_boards.plot.barh("Player", "TRB")

ax.set_title("Top 10 Total Rebounders")
ax.set_xlabel("Total Rebounds")
ax.set_ylabel("Player Name")
ax.grid(False)

ax.get_legend().remove()

for i, v in enumerate(x_labels):
    ax.text(v + 4.5, i - 0.25, str(v), color='royalblue')
```

```python
pick_pocket = stats_df[stats_df["G"]> 70].sort_values("STL", ascending = False).head(10)
pick_pocket
```
